### 1. Scrape the details of most viewed videos on YouTube from Wikipedia. 
- Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:
- Rank
- Name
- Artist
- Upload date
- Views

```python
driver = webdriver.Chrome()
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
Rank = []
Name = []
Artist = []
Upload_date = []
Views = []

rank = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
video_name = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
artist = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')
upload_date = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]')
views = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]')
```

```python
for i in range(len(rank)):
    Rank.append(rank[i].text)
    Name.append(video_name[i].text)
    Artist.append(artist[i].text)
    Upload_date.append(upload_date[i].text)
    Views.append(views[i].text)
```

```python
df = pd.DataFrame({'Rank':Rank, 'Video_name': Name, 'Uploader': Artist,'Published_date':Upload_date, 'Views': Views})
df
```

```python
driver.quit()
```

### 6. Scrape the details of Highest selling novels.
- Url = "https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare"
- You have to find the following details:
- compare
- Book name
- Author name
- Volumes sold
- Publisher
- Genre

```python
# Import selenium webdriver
from selenium import webdriver

# Create a driver object
driver = webdriver.Chrome()

# Navigate to the url
url = "https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare"
driver.get(url)

# Find the table element by xpath
table = driver.find_element(By.XPATH,"//table[@class='in-article sortable']")

# Find all the rows in the table by xpath
rows = table.find_elements(By.XPATH,"//tr")

# Loop through the rows and extract the details
for row in rows:
    # Find the book name by xpath
    book_name = row.find_element(By.XPATH,"//td[1]").text
    
    # Find the author name by xpath
    author_name = row.find_element(By.XPATH,"//td[2]").text
    
    # Find the volumes sold by xpath
    volumes_sold = row.find_element(By.XPATH,"//td[3]").text
    
    # Find the publisher by xpath
    publisher = row.find_element(By.XPATH,"//td[4]").text
    
    # Find the genre by xpath
    genre = row.find_element(By.XPATH,"//td[5]").text
    
    # Print the details
    print(f"Book name: {book_name}")
    print(f"Author name: {author_name}")
    print(f"Volumes sold: {volumes_sold}")
    print(f"Publisher: {publisher}")
    print(f"Genre: {genre}")
    print()

# Close the driver
driver.close()
```

### 7. Scrape the details most watched tv series of all time from imdb.com.
- Url = https://www.imdb.com/list/ls095964455/ You have to find the following details:
- Name
- Year span
- Genre
- Run time
- Ratings
- Votes

```python
# Import selenium webdriver
from selenium import webdriver

# Create a driver object
driver = webdriver.Chrome()

# Navigate to the url
url = "https://www.imdb.com/list/ls095964455/"
driver.get(url)

# Find the list element by xpath
list = driver.find_element(By.XPATH,"//div[@class='lister list detail sub-list']")

# Find all the items in the list by xpath
items = list.find_elements(By.XPATH,"//div[@class='lister-item mode-detail']")

# Loop through the items and extract the details
for item in items:
    # Find the name by xpath
    name = item.find_element(By.XPATH,"//h3[@class='lister-item-header']/a").text
    
    # Find the year span by xpath
    year_span = item.find_element(By.XPATH,"//h3[@class='lister-item-header']/span[2]").text
    
    # Find the genre by xpath
    genre = item.find_element(By.XPATH,"//p[@class='text-muted text-small']/span[5]") .text
    
    # Find the run time by xpath
    run_time = item.find_element(By.XPATH,"//p[@class='text-muted text-small']/span[3]") .text
    
    # Find the ratings by xpath
    ratings = item.find_element(By.XPATH,"//div[@class='ipl-rating-star small']/span[2]") .text
    
    # Find the votes by xpath
    votes = item.find_element(By.XPATH,"//p[@class='text-muted text-small']/span[2]") .text
    
    # Print the details
    print(f"Name: {name}")
    print(f"Year span: {year_span}")
    print(f"Genre: {genre}")
    print(f"Run time: {run_time}")
    print(f"Ratings: {ratings}")
    print(f"Votes: {votes}")
    print()

# Close the driver
driver.close()
```

### 8. Details of Datasets from UCI machine learning repositories.
- Url = https://archive.ics.uci.edu/
- You have to find the following details:
-  Dataset name
- Data type
- Task
- Attribute type
- No of instances
- No of attribute G) Year
- Note: - from the home page you have to go to the Show All Dataset page through code.

```python
# Import the required modules
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Create a webdriver instance
driver = webdriver.Chrome()

# Navigate to the UCI home page
driver.get("https://archive.ics.uci.edu/")

# Find and click on the "View ALL Data Sets" link
link = driver.find_element(By.XPATH, '//input[@class="input-bordered input input-sm join-item w-full"]')
link.click()

# Wait for the table of datasets to load
table = driver.find_element(By.XPATH, '//a[@class="btn-primary btn"]')

# Find all the rows in the table body
rows = table.find_elements(By.XPATH,'//a[@class="w-full truncate"]')

# Loop through each row and extract the details
for row in rows:
    # Find all the cells in the row
    cells = row.find_elements(By.XPATH,'//span[@class="swap-on text-primary-content"]')
    cells.click()
    
    # Get the dataset name from the first cell
    dataset_name = cells.find_element(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"][0]').text
    # Get the data type from the second cell
    data_type = cells.find_element(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"][1]').text
    # Get the task from the third cell
    task = cells.find_element(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"][2]').text
    # Get the attribute type from the fourth cell
    attribute_type = cells.find_element(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"][3]').text
    # Get the number of instances from the fifth cell
    no_of_instances = cells.find_element(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"][4]').text
    # Get the number of attributes from the sixth cell
    no_of_attributes = cells.find_element(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"][5]').text
    # Get the year from the seventh cell
    year = cells.find_element(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"][6]').text

    # Print the details in a formatted way
    print(f"Dataset name: {dataset_name}")
    print(f"Data type: {data_type}")
    print(f"Task: {task}")
    print(f"Attribute type: {attribute_type}")
    print(f"No of instances: {no_of_instances}")
    print(f"No of attributes: {no_of_attributes}")
    print(f"Year: {year}")
    print("-"*50)

# Close the driver
driver.close()
```
