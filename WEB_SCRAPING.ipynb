### 3) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape and make data frame-
- a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
- b) Top 10 ODI Batsmen along with the records of their team andrating.
- c) Top 10 ODI bowlers along with the records of their team andrating.

### a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
response = requests.get(url)
html_content = response.text
soup = BeautifulSoup(html_content, "html.parser")
table = soup.find("table", class_="table")
teams = []
matches = []
points = []
rating = []

for row in table.find_all("tr")[1:11]:  
    cols = row.find_all("td")
    teams.append(cols[1].text.strip())
    matches.append(int(cols[2].text))
    points.append(int(cols[3].text.replace(',', ''))) 
    rating.append(int(cols[4].text))
data = {
    "Team": teams,
    "Matches": matches,
    "Points": points,
    "Rating": rating
}

df = pd.DataFrame(data)

df
```

#### 5) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame-
- i) Headline
- ii) Time
- iii) News Link

### i) Headline

```python
url = "https://www.cnbc.com/world/?region=world"
response = requests.get(url)
html_content = response.text
soup = BeautifulSoup(html_content, "html.parser")
headlines = []
for headline in soup.find_all("h3", class_="Card-title"):
    headlines.append(headline.text.strip())
data = {"Headline": headlines}
df = pd.DataFrame(data)
print(df)
```

### ii) Time

```python
url = "https://www.cnbc.com/world/?region=world"
response = requests.get(url)
html_content = response.text
soup = BeautifulSoup(html_content, "html.parser")
timestamps = []
for timestamp in soup.find_all("time"):
    timestamps.append(timestamp.text.strip())
data = {"Time": timestamps}
df = pd.DataFrame(data)
df
```

### iii) News Link

```python
url = "https://www.cnbc.com/world/?region=world"
response = requests.get(url)
html_content = response.text
soup = BeautifulSoup(html_content, "html.parser")
links = []
for link in soup.find_all("a", class_="Card-titleLink"):
    links.append(link["href"])
data = {"News Link": links}
df = pd.DataFrame(data)
print(df)
```

#### 6) Write a python program to scrape the details of most downloaded articles from AI in last 90 days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles Scrape below mentioned details and make data frame-
- i) Paper Title
- ii) Authors
- iii) Published Date
- iv) Paper URL

### i) Paper Title

```python
url = "https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles"
response = requests.get(url)
html_content = response.text
soup = BeautifulSoup(html_content, "html.parser")
paper_titles = []
for title in soup.find_all("h2", class_="pod-list-title"):
    paper_titles.append(title.text.strip())
data = {"Paper Title": paper_titles}
df = pd.DataFrame(data)
print(df)
```

### ii) Authors

```python
url = "https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles"
response = requests.get(url)
html_content = response.text
soup = BeautifulSoup(html_content, "html.parser")
authors_list = []
for authors in soup.find_all("ul", class_="author-list"):
    authors_names = [author.text.strip() for author in authors.find_all("li")]
    authors_list.append(authors_names)
data = {"Authors": authors_list}
df = pd.DataFrame(data)
print(df)
```

### iii) Published Date

```python
url = "https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles"
response = requests.get(url)
html_content = response.text
soup = BeautifulSoup(html_content, "html.parser")
dates = []
for date in soup.find_all("span", class_="pod-list-content-date"):
    dates.append(date.text.strip())
data = {"Published Date": dates}
df = pd.DataFrame(data)
print(df)
```

### iv) Paper URL

```python
url = "https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles"
response = requests.get(url)
html_content = response.text
soup = BeautifulSoup(html_content, "html.parser")
paper_urls = []
for link in soup.find_all("a", class_="pod-list-title-link"):
    paper_urls.append(link["href"])
data = {"Paper URL": paper_urls}
df = pd.DataFrame(data)
print(df)
```

#### 7) Write a python program to scrape mentioned details from dineout.co.in and make data frame-
- i) Restaurant name
- ii) Cuisine
- iii) Location
- iv) Ratings
- v) Image URL

### i) Restaurant name

```python
url = "https://www.dineout.co.in/delhi-restaurants"
response = requests.get(url)
html_content = response.text
soup = BeautifulSoup(html_content, "html.parser")
restaurant_names = []
for name in soup.find_all("div", class_="restnt-info"):
    restaurant_name = name.find("h3")
    if restaurant_name:
        restaurant_names.append(restaurant_name.text.strip())
data = {"Restaurant Name": restaurant_names}
df = pd.DataFrame(data)
print(df)
```

### ii) Cuisine

```python
url = "https://www.dineout.co.in/delhi-restaurants"
response = requests.get(url)
html_content = response.text
soup = BeautifulSoup(html_content, "html.parser")
cuisines = []
for cuisine in soup.find_all("p", class_="tags-cuisine"):
    cuisines.append(cuisine.text.strip())
data = {"Cuisine": cuisines}
df = pd.DataFrame(data)
print(df)
```

### iii) Location

```python
url = "https://www.dineout.co.in/delhi-restaurants"

# Fetch the HTML content of the page
response = requests.get(url)
html_content = response.text

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")

# Find all the location details
locations = []
for location in soup.find_all("span", class_="restnt-loc"):
    locations.append(location.text.strip())

# Create a DataFrame from the extracted data
data = {"Location": locations}
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

### iv) Ratings

```python
url = "https://www.dineout.co.in/delhi-restaurants"
response = requests.get(url)
html_content = response.text
soup = BeautifulSoup(html_content, "html.parser")
ratings = []
for rating in soup.find_all("div", class_="rating-review"):
    rating_value = rating.find("span", class_="rating").text.strip()
    ratings.append(rating_value)
data = {"Ratings": ratings}
df = pd.DataFrame(data)
print(df)
```

### v) Image URL

```python
url = "https://www.dineout.co.in/delhi-restaurants"
response = requests.get(url)
html_content = response.text
soup = BeautifulSoup(html_content, "html.parser")
image_urls = []
for image in soup.find_all("div", class_="restnt-thumbnail"):
    image_url = image.find("img")["src"]
    image_urls.append(image_url)
data = {"Image URL": image_urls}
df = pd.DataFrame(data)
print(df)
```
