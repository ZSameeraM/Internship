```python
product_urls = []
start=0
end=3
for page in range(start,end):
    urls = driver.find_elements(By.XPATH,'//a[@class = "a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for url in urls:
        product_urls.append(url.get_attribute("href"))
    nxt_button = driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    nxt_button.click()
    time.sleep(2)    
```

```python
Brand=[]
Name=[]
Price=[]
RE=[]
ED=[]
Avail=[]
```

```python
for url in product_urls:
    driver.get(url)
    time.sleep(2)
    
    try:
        brand = driver.find_element(By.XPATH,"/html/body/div[4]/div[2]/div[3]/div[11]/div[47]/div/div[1]/div/table/tbody/tr[1]/td[2]/span")
        Brand.append(brand.text)
    except NoSuchElementException:
        Brand.append('-')
        
    try:
        name = driver.find_element(By.XPATH,'//td[@class="a-span9"]')
        Name.append(name.text)
    except NoSuchElementException:
        Name.append("-")
        
    try:
        price = driver.find_element(By.XPATH,"//span[@id='priceblock_ourprice']")
        Price.append(price.text)
    except NoSuchElementException:
        Price.append("-")
        
    try:
        return_exchange = driver.find_element(By.XPATH,"//div[@id='RETURNS_POLICY']")
        RE.append(return_exchange.text)
    except NoSuchElementException:
        RE.append("-")
        
    try:
        expected_delivery = driver.find_element(By.XPATH,"//div[@id='dynamicDeliveryMessage']")
        ED.append(expected_delivery.text)
    except NoSuchElementException:
        ED.append("-")        
        
    try:
        availability = driver.find_element(By.XPATH,"//div[@id='availability']")
        Avail.append(availability.text)
    except NoSuchElementException:
        Avail.append("-")
            
driver.quit()
```

```python
df = pd.DataFrame({
    "Brand": Brand,
    "Name": Name,
    "Price": Price,
    "Return/Exchange": RE,
    "Expected Delivery": ED,
    "Availability": Avail    
})

# Save to CSV
df.to_csv("macbook_air_products_xpath.csv", index=False)

print("Scraping and saving complete.")
```

#### 5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.


```python
from selenium import webdriver
from time import sleep
import re
```

```python
driver = webdriver.Chrome()
driver.get("https://www.google.co.in/maps/")
```

```python
search_box = driver.find_element(By.ID,"searchboxinput")
search_box.send_keys("Vijayanagar colony")
search_box.send_keys(Keys.RETURN)
time.sleep(5)  

try:
    url_string = driver.current_url
    print('URL Extracted: ',url_string)
    lat_lng = re.findall(r'@(.*)data',url_string)
    
    if lat_lng:
        lat_lng = lat_lng[0].split(',')
        latitude = lat_lng[0]
        longitude = lat_lng[1]
        print('Latitude:', latitude)
        print('Longitude:', longitude)
    else:
        print('Latitude and longitude not found in the URL.')
except Exception as e:
    print('An error occurred:', str(e))

# Close the WebDriver
driver.quit()
```

#### 7. Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped: “Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”.


```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
url = "https://www.forbes.com/billionaires/"
driver.get(url)
time.sleep(5)  
```

```python
# Find all the billionaire listings on the page
billionaire_listings = driver.find_elements(By.XPATH,"//div[contains(@class, 'personCard')]")

# Initialize lists to store billionaire details
ranks = []
names = []
net_worths = []
ages = []
citizenships = []
sources = []
industries = []

# Loop through each billionaire listing
for billionaire in billionaire_listings:
    rank = billionaire.find_element(By.XPATH,".//div[@class='rank']").text
    name = billionaire.find_element(By.XPATH,".//div[@class='name']").text
    net_worth = billionaire.find_element(By.XPATH,".//div[@class='netWorth']").text
    age = billionaire.find_element(By.XPATH,".//div[@class='age']").text
    citizenship = billionaire.find_element(By.XPATH,".//div[@class='countryOfCitizenship']").text
    source = billionaire.find_element(By.XPATH,".//div[@class='source']").text
    industry = billionaire.find_element(By.XPATH,".//div[@class='category']").text

    ranks.append(rank)
    names.append(name)
    net_worths.append(net_worth)
    ages.append(age)
    citizenships.append(citizenship)
    sources.append(source)
    industries.append(industry)

# Close the WebDriver
driver.quit()
```

```python
# Create a DataFrame
df = pd.DataFrame({
    "Rank": ranks,
    "Name": names,
    "Net Worth": net_worths,
    "Age": ages,
    "Citizenship": citizenships,
    "Source": sources,
    "Industry": industries
})
```

```python
# Save to CSV
df.to_csv("billionaires_forbes.csv", index=False)

print("Scraping and saving complete.")
```

### 8. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted from any YouTube Video.


```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
```

```python
driver = webdriver.Chrome()

# URL of the YouTube video
video_url = "https://www.youtube.com/watch?v=YOUR_VIDEO_ID"

# Navigate to the video URL
driver.get(video_url)
time.sleep(5) 
```

```python
# Scroll down to load more comments
scroll_pause_time = 2
scrolls = 10  # Adjust this value to load more comments
for _ in range(scrolls):
    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
    time.sleep(scroll_pause_time)
```

```python
# Find all the comment elements
comment_elements = driver.find_elements(By.XPATH,"//yt-formatted-string[@id='content-text']")
```

```python
# Find all the upvote elements
upvote_elements = driver.find_elements(By.XPATH,"//span[@id='vote-count-middle']")
```

```python
# Find all the timestamp elements
timestamp_elements = driver.find_elements(By.XPATH,"a[@class='yt-simple-endpoint style-scope yt-formatted-string']")
```

```python
# Initialize lists to store comment data
comments = []
upvotes = []
timestamps = []

# Extract comment text, upvotes, and timestamps
for comment_element, upvote_element, timestamp_element in zip(comment_elements, upvote_elements, timestamp_elements):
    comment = comment_element.text
    upvote = upvote_element.text if upvote_element else "0"
    timestamp = timestamp_element.text
    
    comments.append(comment)
    upvotes.append(upvote)
    timestamps.append(timestamp)

# Close the WebDriver
driver.quit()

# Create a DataFrame
df = pd.DataFrame({
    "Comment": comments,
    "Upvotes": upvotes,
    "Timestamp": timestamps
})

# Save to CSV
df.to_csv("youtube_comments.csv", index=False)

print("Scraping and saving complete.")
```

#### 9. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape
hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description.

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
```

```python
driver = webdriver.Chrome()
# URL of the page with hostels in London
url = "https://www.hostelworld.com/search?search_keywords=London,%20England&country=England&city=London&date_from=2023-08-29&date_to=2023-09-10"

# Navigate to the URL
driver.get(url)
```

```python

# Find all the hostel listings on the page
hostel_listings = driver.find_elements(By.XPATH,"//div[@class='result']")

# Initialize a list to store hostel details
hostel_details = []

# Loop through each hostel listing
for hostel in hostel_listings:
    name = hostel.find_element(By.XPATH,".//h2").text.strip()
    distance = hostel.find_element(By.XPATH,".//span[@class='distance']").text.strip()
    rating = hostel.find_element(By.XPATH,".//div[@class='score orange']").text.strip()
    reviews = hostel.find_element(By.XPATH,".//div[@class='reviews']").text.strip()
    privates_price = hostel.find_element(By.XPATH,".//div[@class='price title-5']").text.strip()
    dorms_price = hostel.find_element(By.XPATH,".//div[@class='price title-6']").text.strip()
    facilities = ", ".join([item.text.strip() for item in hostel.find_elements(By.XPATH,".//div[@class='facilities-icons']")])
    description = hostel.find_element(By.XPATH,".//p[@class='property-description']").text.strip()

    hostel_details.append({
        "Hostel Name": name,
        "Distance from City Centre": distance,
        "Rating": rating,
        "Total Reviews": reviews,
        "Privates From Price": privates_price,
        "Dorms From Price": dorms_price,
        "Facilities": facilities,
        "Property Description": description
    })

# Close the WebDriver
driver.quit()

# Create a DataFrame
df = pd.DataFrame(hostel_details)

# Save to CSV
df.to_csv("hostels_london.csv", index=False)

print("Scraping and saving complete.")
```
